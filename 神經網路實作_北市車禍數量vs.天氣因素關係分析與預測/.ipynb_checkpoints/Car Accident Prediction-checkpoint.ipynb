```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("108年臺北市道路交通事故斑點圖.csv")
df
```

```python
df['區']=df['地址'].str[1:4]
df['日期']=df['時間'].str[:-7]
df['日期']=df['日期'].str[1:]
df['時間']=df['時間'].str[1:-1]
df['時間']=df['時間'].str.replace("-"," ")
df['時間'] = pd.to_datetime(df['時間'], format ='%Y-%m-%d %H:%M:%S')
df['日期'] = pd.to_datetime(df['日期'], format ='%Y-%m-%d')
```

```python
#replace trace amount T to 0
df['降水量(mm)']=df['降水量(mm)'].str.replace('T','0')
df['最大六十分鐘降水量(mm)']=df['最大六十分鐘降水量(mm)'].str.replace('T','0')
df[['降水量(mm)', '最大六十分鐘降水量(mm)']] = df[['降水量(mm)', '最大六十分鐘降水量(mm)']].apply(pd.to_numeric)
```

```python
#date encoding
df['年']=df['時間'].dt.year
df['月']=df['時間'].dt.month
df['日']=df['時間'].dt.day
df['時']=df['時間'].dt.hour
df['分']=df['時間'].dt.minute
```

```python
target_columns=['月','日','時','分','X','Y','氣溫(℃)','最高氣溫(℃)','最低氣溫(℃)','相對溼度(%)',
         '風速(m/s)','最大陣風(m/s)','降水量(mm)','降水時數(hour)','最大六十分鐘降水量(mm)','能見度(km)','總雲量(0~10)']
```

```python
#using KNN to address missing value
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier()
humi_test=df[target_columns][df[target_columns]['相對溼度(%)']=='X'].drop('相對溼度(%)',axis=1)
humi_train_x=df[target_columns][df[target_columns]['相對溼度(%)']!='X'].drop('相對溼度(%)',axis=1)
humi_train_y=df[target_columns][df[target_columns]['相對溼度(%)']!='X']['相對溼度(%)']
classifier.fit(humi_train_x,humi_train_y)
pred_y=classifier.predict(humi_test)
```

```python
humi_missing_index=df[df['相對溼度(%)']=='X'].index
for i in range(len(humi_missing_index)):
    df.at[humi_missing_index[i],'相對溼度(%)']=pred_y[i]
```

```python
from sklearn.preprocessing import MinMaxScaler
cols_to_scale=['X','Y','氣溫(℃)','最高氣溫(℃)','最低氣溫(℃)','相對溼度(%)',
         '風速(m/s)','最大陣風(m/s)','降水量(mm)','降水時數(hour)','最大六十分鐘降水量(mm)','能見度(km)','總雲量(0~10)']
scaler=MinMaxScaler()
df[cols_to_scale]=scaler.fit_transform(df[cols_to_scale])
```

```python
#降水量T(微量), 最大六十分鐘降水量T, 相對溼度X(故障)
data=df[target_columns]
labels=df['處理別'].astype(str)
labels=labels.str.replace('1','0')
labels=labels.str.replace('2','1')
labels=labels.str.replace('3','2')
```

```python
df.columns
```

```python
from sklearn.model_selection import train_test_split
# x_train,x_test, y_train, y_test=train_test_split(data,labels, test_size=0.2,random_state=5)
x_train,x_test, y_train_cat, y_test_cat=train_test_split(data,labels, test_size=0.2,random_state=5)

#onehot encoding
from tensorflow.keras.utils import to_categorical
y_train=to_categorical(y_train_cat)
y_test=to_categorical(y_test_cat)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense 
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy

#create neural network
model=Sequential([
    keras.layers.Dense(256,input_shape=(17,),activation='relu'),
    keras.layers.Dense(128,activation='relu'),
    keras.layers.Dense(64,activation='relu'),
    keras.layers.Dense(32,activation='relu'),
    keras.layers.Dense(3,activation='sigmoid')
])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
print(pd.DataFrame(x_train))
print(y_train)
```

```python
#training progress
model.fit(x_train,y_train, epochs=50, batch_size=64)
```

```python
yp=model.predict(x_test)
yp
```

```python
result=[]
for i in yp:
    index = np.where(i == np.amax(i))
    result.append(index[0][0])
```

```python
y_test_cat=list(y_test_cat)
y_test_cat=[int(i) for i in y_test_cat]
```

```python
#check the precision, recall and f1-score
from sklearn.metrics import confusion_matrix, classification_report
print(classification_report(y_test_cat,result))
```
