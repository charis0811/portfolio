```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel('108年臺北市道路交通事故斑點圖.xlsm', index_col=[0])
df = df.loc[ : , ~df.columns.str.contains("Unnamed")]
df.head(5)
```

```python
print('There are a total of {} rows and {} columns in the original dataset'.format(df.shape[0],df.shape[1]))
print("Are there null values in the original dataset?: {}".format(df.isnull().values.any()))
print('Column names in the dataset are: \n{}'.format(('\n').join(df.columns)))
```

```python
# Data types
column_datatypes = set()
for column in df.columns:
    column_datatypes.add(str(df[column].dtype))
print("The dataset contains {} different data types and they are: {}".format(len(column_datatypes), ", ".join(column_datatypes)))
```

```python
#df=df[df["Age"]<=33].index
X = df.drop(columns='嚴重度')
y = df['嚴重度']
```

```python
numerical_features = list()
categorical_features = list()
for column in X.columns:
    # In the dataset we only have float and int64.
    if (df[column].dtype == 'float64' or df[column].dtype == 'int64'):
        numerical_features.append(column)
    # Categorical
    elif (df[column].dtype == 'object'):
        categorical_features.append(column)
```

```python
print('There are a total of {} numerical features in the dataset.'.format(len(numerical_features)))
```

```python
X_num_total = X[numerical_features]
```

```python
X_num_total.hist(bins=60,figsize=(20, 10))
plt.show()
#各attribute與嚴重度的關係
```

```python
# Checking the correlation between the numerical features
plt.subplots(figsize=(30,16))
sns.heatmap(X_num_total.corr(), cmap="YlGnBu", annot=True)
plt.title('Correlation Matrix')
plt.show()
```
