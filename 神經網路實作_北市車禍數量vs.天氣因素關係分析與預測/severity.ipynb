```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel('108年臺北市道路交通事故斑點圖.xlsm', index_col=[0])
df = df.loc[ : , ~df.columns.str.contains("Unnamed")]
df.head(5)
```

```python
for i in range(0, df.shape[0]):
    df['地址'][i] = df['地址'][i].split("區")[0]
df.head(10)
```

```python
severity = df.groupby("嚴重度")
print("severity的資料類型為", type(severity))
```

```python
print("每個資料量的大小分別為")
print(severity.size())
```

```python
severity1 = severity.get_group(1)
severity1
```

```python
address1 = severity1.groupby('地址')
#address1.size()
#print(type(address1))
count_1 = severity1['地址'].value_counts()
print("嚴重度1的各區車禍數量")
#count_1 = address1.to_frame()
print(count_1)
```

```python
X = severity1.drop(columns='嚴重度')
y = severity1['嚴重度']
numerical_features = list()
categorical_features = list()
for column in X.columns:
    # In the dataset we only have float and int64.
    if (severity1[column].dtype == 'float64' or severity1[column].dtype == 'int64'):
        numerical_features.append(column)
    # Categorical
    elif (severity1[column].dtype == 'object'):
        categorical_features.append(column)
print('There are a total of {} numerical features in the dataset.'.format(len(numerical_features)))
```

```python
X_num_total = X[numerical_features]
X_num_total.hist(bins=60,figsize=(20, 10))
plt.show()
#各attribute與嚴重度1的關係
```

```python
# Checking the correlation between the numerical features
plt.subplots(figsize=(30,16))
sns.heatmap(X_num_total.corr(), cmap="YlGnBu", annot=True)
plt.title('Correlation Matrix')
plt.show()
```
